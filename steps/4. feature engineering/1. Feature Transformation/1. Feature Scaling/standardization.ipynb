```python
import pandas as pd

# Sample student data
df = pd.DataFrame({
    'Age': [18, 19, 17, 21, 20],
    'ExamScore': [88, 92, 85, 70, 76],
    'FamilyIncome': [30000, 45000, 28000, 52000, 61000]
})

print("Original Data:")
print(df)
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit and transform the data
scaled_data = scaler.fit_transform(df)

# Convert back to a DataFrame
df_scaled = pd.DataFrame(scaled_data, columns=df.columns)

print("\nStandardized Data:")
print(df_scaled)
```

```python
# Plotting before and after standardization
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 5))

# Before Standardization
plt.subplot(1, 2, 1)
plt.scatter(df['FamilyIncome'], df['ExamScore'], color='blue')
plt.title("Before Standardization")
plt.xlabel("Family Income")
plt.ylabel("Exam Score")

# After Standardization
plt.subplot(1, 2, 2)
plt.scatter(df_scaled['FamilyIncome'], df_scaled['ExamScore'], color='green')
plt.title("After Standardization")
plt.xlabel("Family Income (standardized)")
plt.ylabel("Exam Score (standardized)")

plt.tight_layout()
plt.show()
```

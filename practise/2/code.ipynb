```python
import pandas as pd
import numpy as np

df = pd.read_csv("train.csv")

print(df.columns)
print(df.head(2))
```

# 1. Categorical Data

a. CountPlot

```python
import seaborn as sns

print(df['Pclass'].value_counts().plot(kind='bar')) # using pandas

sns.countplot(x=df["Sex"])  # using seaborn
```

b.pie chart

```python
print(df['Sex'].value_counts().plot(kind='pie', autopct="%.2f") )
```

# 2. Numerical Data

2.a Histogram

```python
'''
Only work for numerical data, help to see the distribition of the data. We are plotting the in some range.count
'''

import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=50)  # bins : helps to see the distribution in small things
```

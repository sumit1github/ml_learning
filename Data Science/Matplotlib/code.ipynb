Bar Plot

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

df = pd.read_csv('train.csv')
x = df['train_number'].to_list()
y = df['speed'].to_list()

new_x = [str(i) for i in x]

plt.xlabel("train no", fontsize=15)
plt.ylabel("speed", fontsize=15)
plt.title("Bar chart", fontsize=15)

color = ["red","green","blue","purple"]
assign_color_list=[]

for i in range(len(x)-1):
    assign_color_list.append(random.choice(color))

plt.bar(new_x,y, width=0.4, color=assign_color_list, align="edge", edgecolor="red", linewidth=5, linestyle=":")
plt.show()
```

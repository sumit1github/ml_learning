```python
from sklearn import datasets, linear_model
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# Step 1: Load the data
df = pd.read_csv("multiple_linear_regression_data.csv")

# Step 2: Features (X) and target (y)
features = ['cgpa', 'internships', 'projects', 'backlogs']
target = 'lpa'
colors = ['red', 'blue', 'green', 'orange']

# Step 3: Plotting each feature vs LPA
plt.figure(figsize=(12, 8))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    plt.scatter(df[feature], df[target], color=colors[i])
    plt.title(f'{feature.capitalize()} vs LPA')
    plt.xlabel(feature.capitalize())
    plt.ylabel('LPA')
    plt.grid(True)

plt.tight_layout()
plt.show()

```

```python
from sklearn.linear_model import LinearRegression


features = ['cgpa', 'internships', 'projects', 'backlogs'] # X
target = 'lpa'   # Y

x = df[features]
y = df[target]

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)

# Initialize the model
model = LinearRegression()

# Train the model
model.fit(x_train, y_train)
```

```python
# Predict using the test data
predictions = model.predict(x_test)

# Score the model (R^2 score)
print(f"Model Accuracy (R score): {model.score(x_test, y_test):.2f}")
```

```python
def predict_cgpa(cgpa, internships, projects, backlogs):
    input_data_frame = pd.DataFrame({
        'cgpa': [cgpa],
        'internships': [internships],
        'projects': [projects],
        'backlogs': [backlogs]
    })
    prediction = model.predict(input_data_frame)
    return max(0, round(prediction[0], 2))

result = predict_cgpa(1.5, 2, 3, 5)
print(f"His Predicted CGPA is : {result}")
```

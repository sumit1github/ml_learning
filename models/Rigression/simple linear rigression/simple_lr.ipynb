```python
from sklearn import datasets, linear_model
import numpy as np
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


df = pd.read_csv('cgpa.csv')


x = df[["cgpa"]]
y = df['lpa']

# plotting
plt.scatter(x, y, color="red")

plt.title("Study Hours vs Marks")
plt.xlabel("cgpa ---->")
plt.ylabel("package ---->")
plt.grid(True)

plt.show()
```

```python


x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)

lr = linear_model.LinearRegression()

lr.fit(x_train,y_train)
```

```python
def predict_package(cgpa_input):
    cgpa_df = pd.DataFrame({'cgpa': [cgpa_input]})
    prediction = lr.predict(cgpa_df)
    return max(0, round(prediction[0], 2))


# Take user input
user_cgpa = float(input("Enter your CGPA: "))
predicted_lpa = predict_package(user_cgpa)
print(f"Predicted Package (LPA) for CGPA : {user_cgpa}: ₹ {predicted_lpa} LPA")
```

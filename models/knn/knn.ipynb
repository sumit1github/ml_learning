```python
# 2. Select Numeric Features (you can adjust based on dataset)
features = ['popularity', 'vote_average', 'vote_count', 'runtime']
movie_features = df[features]
```

```python
# Drop rows with any NaN values in selected features
movie_features = df[features].dropna()
```

```python
from sklearn.preprocessing import StandardScaler

# 3. Normalize Features
scaler = StandardScaler()
scaled_features = scaler.fit_transform(movie_features)
```

```python
# 4. Fit KNN Model
from sklearn.neighbors import NearestNeighbors

knn = NearestNeighbors(n_neighbors=6, algorithm='auto', metric='euclidean')
knn.fit(scaled_features)
```

```python
# 5. Pick a Movie (e.g., index 0)
movie_index = 0
distances, indices = knn.kneighbors([scaled_features[movie_index]])
```

```python
# 6. Show Similar Movies
print("Query Movie:", df['title'][movie_index])
print("\nRecommended Movies:")
for idx in indices[0][1:]:  # skip the first one (itself)
    print(df['title'][idx])
```

```python
# Function to recommend movies

import matplotlib.pyplot as plt

def recommend_movies(movie_title):
    # Check if movie exists
    if movie_title not in df['title'].values:
        print("Movie not found in dataset.")
        return
    
    movie_index = df[df['title'] == movie_title].index[0]
    distances, indices = knn.kneighbors([scaled_features[movie_index]])

    print(f"\nüé¨ Query Movie: {movie_title}")
    print("\nüìΩÔ∏è Recommended Movies:\n")
    recommended_titles = []
    recommended_distances = []
    
    for i, idx in enumerate(indices[0][1:]):  # skip the input movie
        title = df['title'][idx]
        dist = distances[0][i+1]
        recommended_titles.append(title)
        recommended_distances.append(dist)
        print(f"{i+1}. {title} (Distance: {dist:.2f})")

    # Plot bar graph
    plt.figure(figsize=(10, 5))
    plt.barh(recommended_titles, recommended_distances, color='skyblue')
    plt.xlabel('Distance (lower = more similar)')
    plt.title(f'Movies similar to "{movie_title}"')
    plt.gca().invert_yaxis()  # Highest similarity on top
    plt.grid(True)
    plt.tight_layout()
    plt.show()

# üîç Example usage
recommend_movies("Inception")
```
